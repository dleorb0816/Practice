  - 제조회사 : audi, toyota 
  - 어느 회사의 도시(cty)연비 평균이 높은지 알아보세요.

```python
# 판다르소 mpg.csv파일을 mpg에 할당한다.
import pandas as pd
mpg = pd.read_csv(r'C:\Users\601\Desktop\김경섭\데이터베이스\데이터가공\mpg.csv')
```

```python
mpg.head()
```

```python
# query로 manufacturer컬럼의 값이 audi와 toyota인 것을 나누어 각각 할당.
audi = mpg.query('manufacturer=="audi"')
toyota = mpg.query('manufacturer=="toyota"')
```

```python
# 잘 할당 되었다.
toyota.head()
```

#### audi와 toyota의 hwy컬럼의 값의 평균을 출력한다.

```python
print(audi['hwy'].mean())
print(toyota['hwy'].mean())
```
