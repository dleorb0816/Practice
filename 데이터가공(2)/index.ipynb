1. mpg 데이터 복사본 만들어주세요.
  - hwy + cty 합계를 추가해주세요.

```python
import pandas as pd
mpg = pd.read_csv(r'C:\Users\601\Desktop\김경섭\데이터베이스\데이터가공(2)\mpg.csv')
```

```python
mpg.head()
```

```python
df = mpg.copy()
```

```python
df.head()
```

```python
df = df.assign(hcy= lambda x : x['hwy']+x['cty'])
df
```

2. 위 변수를 2로 나눠서 평균 변수 추가해주세요.

```python
df = df.assign(hmean= lambda x : x['hcy']/2)
```

```python
df
```

3. 평균 높은 순으로 정렬해주세요. 가장 높은 3개만 출력

```python
df.sort_values('hmean',ascending=0).head(3)
```

4. 각각의 실행문을 완료 후 한 줄로 만들어주세요.
